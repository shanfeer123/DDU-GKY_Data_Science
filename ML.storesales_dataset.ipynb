```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('./Store Sales.csv',parse_dates=['Order Date','Ship Date'])
```

```python
data.head(10)
```

```python
data.dropna(inplace=True)
```

```python
data.columns
```

```python
data.rename(columns={'Ship Mode':'ShipMode','Customer ID':'CustomerID','Customer Name':'CustomerName','Ship Date':'ShipDate','Sub-Category':'SubCategory','Order Date':'OrderDate','Postal Code':'PostalCode','Product ID':'ProductID','Product Name':'ProductName'},inplace=True)
                
```

```python
data.columns
```

#### Extracting the date 

```python
data["year"]=data.OrderDate.dt.year
data["month"]=data.OrderDate.dt.month
data["Day"]=data.OrderDate.dt.day
```

```python
data["year"]=data.ShipDate.dt.year
data["month"]=data.ShipDate.dt.month
data["Day"]=data.ShipDate.dt.day
```

```python
data.head(10)
```

```python
data.drop(columns=['Row ID','Order ID','CustomerID','PostalCode','ProductID'],inplace=True)
```

```python
data.head(10)
```

```python
data['Ship_Delay'] = data.ShipDate - data.OrderDate
data
```

```python
data['Ship_Delay_01'] = data.Ship_Delay.astype('str')
data
```

```python
data.info()
```

```python
data['Ship_Delay_01'] =data.Ship_Delay_01.str.replace('\D', "", regex=True)
data
```

```python
data.info()
```

```python
data['Ship_Delay_01'] = data.Ship_Delay_01.astype('int')
data
```

```python
data.info()
```

```python
data['WeekDay'] = data.OrderDate.dt.weekday
data
```

```python
data.WeekDay.unique()
```

```python
data.WeekDay.replace({0:'Sunday', 1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday'}, inplace=True)
data
```

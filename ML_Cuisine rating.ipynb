```python
import pandas as pd
import numpy as np
import warnings
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('./Cuisine_rating.csv')
data.head(10)
```

#### Renaming the columns 

```python
data.rename(columns={'Food Rating':'FoodRating','Service Rating':'ServiceRating','Overall Rating':'OverallRating','Alcohol ':'Alcohol'},inplace=True)
                
                     
                    
```

```python
data.isnull().mean()*100
```

```python
data['Age']=[(2022-DOB)for DOB in data['YOB']]
```

```python
data.drop(columns=['User ID','Area code','YOB'],inplace=True)
```

#### Histogram 

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.histplot(data=data,x='FoodRating',bins=5,kde=True)

plt.subplot(2,2,2)
sns.histplot(data=data,x='ServiceRating',bins=5,kde=True)

plt.subplot(2,2,3)
sns.histplot(data=data,x='OverallRating',bins=5,kde=True)

plt.subplot(2,2,4)
sns.histplot(data=data,x='Age',bins=5,kde=True)
plt.show()
```

#### Box plot 

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
sns.boxplot(data=data,x='FoodRating')

plt.subplot(2,2,2)
sns.boxplot(data=data,x='ServiceRating')

plt.subplot(2,2,3)
sns.boxplot(data=data,x='OverallRating')

plt.subplot(2,2,4)
sns.boxplot(data=data,x='Age')
plt.show()
```

#### count plot 

```python
plt.figure(figsize=(22,15))

plt.subplot(4,2,1)
sns.countplot(data=data,x='Gender')

plt.subplot(4,2,2)
sns.countplot(data=data,x='Activity')

plt.subplot(4,2,3)
sns.countplot(data=data,x='Marital Status')

plt.subplot(4,2,4)
sns.countplot(data=data,x='Budget')

plt.subplot(4,2,5)
sns.countplot(data=data,x='Alcohol')

plt.subplot(4,2,6)
sns.countplot(data=data,x='Smoker')

plt.subplot(4,2,7)
sns.countplot(data=data,x='Cuisines')

plt.subplot(4,2,8)
sns.countplot(data=data,x='Location')
plt.xticks(rotation = 90)

plt.show()
```

#### Bar plot 

```python
Food_cuisines = data.groupby(['Cuisines'])['FoodRating'].mean()
Food_cuisines
```

```python
sns.barplot(data=data,x='Cuisines',y='FoodRating',hue='Gender')
plt.legend(bbox_to_anchor=(1.3, 1))
plt.show()
```

```python
def sort_Location(location):
    data_location = data[data.Location==location]
    data_location.drop(columns='Location',inplace=True)
    return data_location
```

```python
Central_Park_NY=sort_Location('Central Park,NY')
```

```python
Central_Park_NY
```

#### Line chart 

```python
plt.figure(figsize=(10,6))
plt.xlabel('OverallRating')
plt.ylabel('Cuisines')
sns.lineplot(data=Central_Park_NY,x='OverallRating',y='Cuisines')
plt.show()
```

#### Pie chart 

```python
def sort_Cuisines(kitchen):
    data_kitchen = data[data.Cuisines==kitchen]
    data_kitchen.drop(columns='Cuisines',inplace=True)
    return data_kitchen
```

```python
kitchen=sort_Cuisines('Japanese')
kitchen
```

```python
data.Activity.value_counts().index
```

```python
palette_color=sns.color_palette('plasma')
```

```python
plt.pie(data=kitchen, x=data.Activity.value_counts(),explode = [0.1,0], colors=palette_color,labels=['Student', 'Professional'],autopct='%0.2f%%');
```

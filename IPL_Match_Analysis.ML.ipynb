#### To get season column 

```python
no_of_matches = data['season'].value_counts().sort_index()
no_of_matches
```

### Univariate Analysis

### Bar plot

```python
no_of_matches.plot(kind='bar' ,color='#bcbd22')
plt.title('Number of Matches Played per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches Played')
plt.show()
```

<font face="Times New Roman" size=3>

So by this graph we get an idea the number of matches played per season

Also, we can conclude that:

In seasons 2011, 2012, 2013, 2022, 2023 , number of matches played are more than the usual range
Rest in almost every season total matches played are nearly equal to 60

#### Toss won

```python
toss_won = data['toss_won'].value_counts().sort_values(ascending=True)
toss_won
```

```python
toss_won.plot(kind='bar',color='purple')
plt.title('Number of Tosses won by each Team')
plt.xlabel('Number of Tosses Won')
plt.ylabel('Team Name')
plt.show()
```

### Pie chart

#### Decision

```python
bowl_first = data['decision'].value_counts()[0]
bat_first = data['decision'].value_counts()[1]
```

```python
plt.pie([bowl_first,bat_first],labels=['Bowl First','Bat First'],autopct='%1.2f%%')
plt.title('Bat First VS Bowl First Ratio')
plt.show()
```

### Histogram

#### Run Distribution

```python
data.home_runs.value_counts().sort_values()
```

```python
runs = data['home_runs'].dropna().astype(int)

plt.figure(figsize=(8, 6))
plt.hist(runs, bins=20, color='orange', edgecolor='black')
plt.title('Run Distribution')
plt.xlabel('Runs')
plt.ylabel('Frequency')
```

### Pie chart

#### Super Over

```python
data['super_over'].value_counts()
```

```python
super_over_matches = data['super_over'].value_counts()

plt.figure(figsize=(8, 6))

plt.pie(super_over_matches.values, labels=super_over_matches.index, autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'])
plt.title('Distribution of Super Over Matches')
plt.show()
```

### Box plot

#### Run rate comparison

```python
data['home_run_rate'] = data['home_runs'] / data['home_overs']
data['away_run_rate'] = data['away_runs'] / data['away_overs']

plt.figure(figsize=(8, 6))

data[['home_run_rate', 'away_run_rate']].plot(kind='box')
plt.title('Run Rate Comparison')
plt.ylabel('Run Rate')
plt.xticks([1, 2], ['Home', 'Away'])
plt.show()
```

###  Total number of boundries by home team and away team per season

```python
boundries = data.groupby('season')[['home_boundaries','away_boundaries']].sum()
boundries
```

```python
type(boundries.index)
```

```python
boundries = boundries.reset_index()
```

```python
boundries
```

### Line plot

```python
boundries.plot.line(x='season')
plt.title('Boundries by Teams in each Season')
plt.show()
```
